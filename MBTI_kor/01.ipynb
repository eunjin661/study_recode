```python
import pandas as pd

dataset = load_dataset("epinfomax/mbti-korean-dataset")

df = pd.DataFrame(dataset["train"])

id2label = {
    0: "ENFJ",
    1: "ENFP",
    2: "ENTJ",
    3: "ENTP",
    4: "ESFJ",
    5: "ESFP",
    6: "ESTJ",
    7: "ESTP",
    8: "INFJ",
    9: "INFP",
    10: "INTJ",
    11: "INTP",
    12: "ISFJ",
    13: "ISFP",
    14: "ISTJ",
    15: "ISTP",
}

df["mbti"] = df["label"].map(id2label)

print(df['text'].head())
```

```python
train_df = pd.DataFrame(dataset["train"])
valid_df = pd.DataFrame(dataset["validation"])
test_df = pd.DataFrame(dataset["test"])

for name, temp_df in {
    "train": train_df,
    "validation": valid_df,
    "test": test_df
}.items():
    temp_df["mbti"] = temp_df["label"].map(id2label)
    print(f"\n[{name}]")
    print(temp_df["mbti"].value_counts().sort_index())
```
