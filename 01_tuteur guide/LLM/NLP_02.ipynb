```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize, RegexpTokenizer
nltk.download('punkt_tab') # sent_tokenize과 word_tokenize을 사용하기 위해 필요한 패키지 다운로드

para = "Hello everyone. It's good to see you. Let's start our text mining class!"
para_kor = "안녕하세요, 여러분. 만나서 반갑습니다. 이제 텍스트마이닝 클래스를 시작해봅시다!"

# 문장 토큰화
sentences = sent_tokenize(para)
sentences_kor = sent_tokenize(para_kor)
print("영문 문장 토큰화:", sentences)
print("한글 문장 토큰화:", sentences_kor)

# 단어 토큰화
words = word_tokenize(para)
words_kor = word_tokenize(para_kor)
print("영문 단어 토큰화:", words)
print("한글 단어 토큰화:", words_kor)

# 정규 표현식을 사용한 단어 토큰화
tokenizer = RegexpTokenizer(r'\w+') # \w+는 문자나 숫자가 1개 이상 반복되는 것을 찾아냄 - 구두점 제거 효과
words_regex = tokenizer.tokenize(para)
words_regex_kor = tokenizer.tokenize(para_kor)
print("영문 정규 표현식 단어 토큰화:", words_regex)
print("한글 정규 표현식 단어 토큰화:", words_regex_kor)
```

```python
# 불용어 제거 (stop words)
from nltk.corpus import stopwords
nltk.download('stopwords') # 불용어 리스트 다운로드
stop_words = set(stopwords.words('english')) # 영어 불용어 리스트 (set - 빠른 검색을 위해, 중복 제거)
text_stop = 'Sorry, I am late. I had to finish my homework.'

# 구두점 제거
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(text_stop.lower()) # 소문자로 변환하여 토큰화

# 기본 불용어 제거
result_nltk = [word for word in tokens if word not in stop_words]
print("불용어 제거 후 토큰:", result_nltk)

# 사용자 정의 불용어 추가
custom_stop_words = set(['sorry', 'homework']) # 사용자 정의 불용어 리스트
result_custom = [word for word in tokens if word not in custom_stop_words]
print("사용자 정의 불용어 제거 후 토큰:", result_custom)
```

```python
# 어간 추출 (stemming) / 표제어 추출 (lemmatization)
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('wordnet') # WordNetLemmatizer를 사용하기 위해 필요한 패키지 다운로드
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

words = ['cooking','cookery','cookbooks','believes']

print(f"{'-'*15} {'-'*15} {'-'*15}")
print(f"{'Original':15} {'Stemming':15} {'Lemmatization':15}")
print(f"{'-'*15} {'-'*15} {'-'*15}")

for word in words:
    stemmed = stemmer.stem(word)
    lemmatized = lemmatizer.lemmatize(word)
    print(f"{word:15} {stemmed:15} {lemmatized:15}")

# 품사(pos)를 지정한 표제어 추출
lammatized_verb = lemmatizer.lemmatize('believes', pos='v') # 동사로 지정하여 표제어 추출
print("동사로 지정한 표제어 추출:", lammatized_verb)
```

```python
# 한국어 텍스트 정규화 및 형태소 분석
# %pip install konlpy JPype1
#  https://download.oracle.com/java/26/latest/jdk-26_windows-x64_bin.exe
# 설치 후 환경변수 JAVA_HOME에 JDK 설치 경로 추가
# C:\Program Files\Java\jdk-26.0.1
from konlpy.tag import Okt
t = Okt()
sentence = '''절망의 반대가 희망은 아니다.\n어두운 밤하늘에 별이 빛나듯\n희망은 절망 속에 싹트는 거지\n만약에 우리가 희망함이 적다면\n그 누가 세상을 비출어줄까.\n정희성, 희망 공부'''
```

```python
print(sentence)
print("형태소 분석:", t.morphs(sentence)) # 형태소 분석
print(f'명사 추출: {t.nouns(sentence)}') # 명사 추출
print(f'품사 태깅: {t.pos(sentence)}') # 품사 태깅
```

```python
# 숫자 정규화
spam_texts = [
    "Cheap iPad for $999! Buy now $799!",
    "Order 123456 shipped on 2026-01-15"
]

import re
for text in spam_texts:
    # 숫자 제거
    text_no_numbers = re.sub(r'\d+', '', text)
    print(f'원문: {text}')
    print(f'숫자 제거 후: {text_no_numbers}')
```
