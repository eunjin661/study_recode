```python
# 상관관계와 산점도 분석

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds.info()
```

```python
diamonds_sample = diamonds.sample(1000,random_state=42)
diamonds_sample.head()
```

```python
# 산점도 / 캐럿이 가격에 미치는 영향
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(diamonds_sample['carat'],diamonds_sample['price'])
ax.set_xlabel('캐럿',fontsize=15)
ax.set_ylabel('캐럿',fontsize=15)
plt.tight_layout()
plt.show()

# 상관계수 계산
corr = diamonds['carat'].corr(diamonds['price'])
print(f'상관계수: {corr:.3f}')

# sns.scatterplot(data=diamonds_sample, ,x='carat', y='price', ax=ax)
```

```python
# regplot 회귀선
# 캐럿을 기준으로 적정 가격 예측

pig, ax = plt.subplots(figsize=(10,6))
sns.regplot(data=diamonds_sample, x='carat', y='price',ax=ax)
```

```python
# 다이아몬드 등급(커팅)에 따라 가격패턴이 다른가?
fig, ax = plt.subplots(2,3,figsize=(8,6))
for idx, cut in enumerate( diamonds_sample['cut'].unique() ):
    sns.regplot(data = diamonds_sample[ diamonds_sample['cut'] == cut ],
                    x='carat',y='price',ax= ax[idx // 3,idx % 3])
    ax[idx // 3,idx % 3].set_title(cut)   
plt.tight_layout()
plt.show()    
```

```python
# 투명도 별로 가격결정 패턴이 다른지

clarity_subset = diamonds_sample[diamonds_sample['clarity'].isin(['IF','VVS1','VS1'])]
g = sns.lmplot(data = clarity_subset, x='carat',y='price', hue='clarity')
plt.tight_layout()
plt.show()
```

```python
# 어떤 속성이 가격에 가장 큰 영향을 미칠까
# 수치형 변수들의 상관계수 계산

numeric_cols = list(diamonds.describe().columns)
numeric_cols.remove('price')
collelation = diamonds[numeric_cols + ['price']].corr()['price'].sort_values()
fig, ax = plt.subplots(figsize=(10,6))
colors = ['blue' if x > 0 else 'red' for x in collelation]
bars = ax.barh(collelation.index, collelation.values,color = colors)

```

```python
# 상관관계 vs 인과관계
# 상관관계는 반드시 인과관계를 의미하지 않음
# 무게와 x 치수의 관계

fig, axes = plt.subplots(1,2,figsize=(16,8))
# 좌측 carat price
sns.regplot(data=diamonds_sample, x='carat', y='price', ax=axes[0], color='blue')
axes[0].set_title('무게 vs 가격')
# 우측 x price
sns.regplot(data=diamonds_sample, x='x', y='price', ax=axes[1], color='red')
axes[1].set_title('치수 vs 가격')

plt.tight_layout()
plt.show()

# 무게가 증가하면 치수도 증가하는 관계 (치수/가격 허위 상관관계)
# 추론이 불필요한 계수 즉 변수 / 머신 러닝 시 학습에서 제외
```
