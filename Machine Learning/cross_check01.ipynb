```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")
    return pd.read_csv(Path("datasets/housing/housing.csv"))

housing = load_housing_data()
```

```python
import matplotlib.pyplot as plt
housing.hist(figsize=(12,8))
plt.show()
```

```python
hists = plt.hist(housing['median_income'], bins=5)
housing['incom_cats'] = pd.cut(housing['median_income'], bins=hists[1], labels=[1,2,3,4,5])
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(10,6))
sns.countplot(data=housing,x='incom_cats',ax=ax, palette='rainbow', hue='incom_cats')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

housing.dropna(inplace=True)

train, test = train_test_split(housing, test_size=0.2, random_state=42,
                               stratify=housing['incom_cats'])
fig, axes = plt.subplots(1,2,figsize=(12,6))
sns.countplot(data=train, x='incom_cats',hue='incom_cats',ax=axes[0])
sns.countplot(data=test, x='incom_cats',hue='incom_cats',ax=axes[1])
plt.tight_layout()
plt.show()
```
